```python
#df = pd.read_excel("Decision_GPT-4_Output_Paula_Lancet.xlsx", header=None, skiprows=1)
df = pd.read_csv('115k_131224_1144_dedup (1).csv') 

df = df[df['decision'] == 'Include']

#ensure all dois are in the same format
df['doi'] = df['doi'].str.replace('http://dx.doi.org/','')
df['doi'] = df['doi'].str.replace('https://dx.doi.org/','')
df['doi'] = df['doi'].str.replace('https://doi.org/','')
df['doi'] = df['doi'].str.replace('http://doi.org/','')
```

```python
# Remove all rows where the 'DOI' column is NaN
df = df.dropna(subset=['doi'])
df_overton = df[['doi']]
df_overton = df_overton.drop_duplicates(subset=['doi'])
df_overton = df_overton.reset_index(drop=True)
# df_overton = df_overton.head(100)
df_overton
```

```python
api_key = "f97559-3dffd1a0a50ba45efc5237f1002aaae7"
base_url = "https://app.overton.io/articles.php"

# data = {
#     'DOI': [
#         '10.25561/77482',
#         '10.1080/02684527.2022.2055936',
#         '10.25561/28409',
#         '10.25561/77299',
#         '10.25561/77344'
#     ]
# }
# df = pd.DataFrame(data)

def get_citations(doi):
    params = {
        "dois": doi,
        "sort": "date",
        "format": "json",
        "api_key": api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if 'results' in data and 'results' in data['results'] and data['results']['results']:
            citations = data['results']['results'][0]['citations']
            cited_by_documents = data['results']['results'][0]['cited_by_documents']
            title = data['results']['results'][0]['title']  # Extract the title
            return citations, cited_by_documents, title
    # Return default values if retrieval fails
    return None, None, None

# Apply the function to each row in the DataFrame
df_overton['Citations'], df_overton['Additional Info'], df_overton['Title'] = zip(*df_overton['doi'].apply(get_citations))

# Flatten the list of additional info and create a new DataFrame
additional_info = []
for _, row in df_overton.iterrows():
    doi = row['doi']
    if row['Additional Info'] is not None:
        for info in row['Additional Info']:
            info['doi'] = doi
            additional_info.append(info)

# Check if additional_info is empty before creating the DataFrame
if additional_info:
    additional_df = pd.DataFrame(additional_info)
    
    # Filter the desired fields if additional_info is not empty
    desired_columns = ['doi', 'source_title', 'document_url', 'document_title', 'country', 'published_on']
    additional_df = additional_df[desired_columns]
    
    # Merge the additional information DataFrame with the original DataFrame
    df_overton = pd.merge(df_overton, additional_df, on='doi', how='left')
else:
    print("No additional info retrieved for any DOI.")

# Drop the 'Additional Info' column
df_overton.drop(columns=['Additional Info'], inplace=True)
df_overton
```
